```python
import google.generativeai as genai

# Configure your Gemini API key
genai.configure(api_key="AIzaSyCqJmDKU58S4qyr45IZIDmBM4mOZrE24UM")

# Initialize the Gemini model
model = genai.GenerativeModel("gemini-2.5-flash")
```

```python
def generate_story(topic):

    prompt = f"""
    Write an engaging and fun story about {topic} for beginners.

    The story should:
    - Be written in a Gen Z style (fun, energetic, and relatable).
    - Use simple and easy-to-understand language.
    - Feel interesting and create good vibes while reading.
    - Explain the topic naturally through the story.
    - Include a few cool or surprising facts.
    - Be around 200–300 words.

    The tone should be friendly, slightly playful, and motivating for someone who is learning this topic for the first time.

    End the story with a short summary of what we learned.
    """

    # Generate content using Gemini
    response = model.generate_content(prompt)

    return response.text


# Take input from user
topic = input("Enter the topic for the story: ")

# Generate story
story = generate_story(topic)

print("\nGenerated Story:\n")
print(story)
```

```python
#Convert the story to speech
#Now that we have generated our story, let's convert it to speech using the gTTS (Google Text-to-Speech) library. We'll create an audio file in memory and play it directly in the notebook using an audio player widget.

#This step may take a while to complete, please be patient.
```

```python

import io
import gtts
from IPython.display import Audio

print("gTTS installed successfully")
```

```python
tts=gtts.gTTS(story)
tts.save("generated_story.mp3")

print("Audio saved as story.mp3")

audio_bytes = io.BytesIO()
tts.write_to_fp(audio_bytes)
audio_bytes.seek(0)

# Create and display an audio player widget in the notebook
Audio(audio_bytes.read(), autoplay=True)
```
